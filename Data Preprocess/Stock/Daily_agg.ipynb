```python
import numpy as np 
import pandas as pd 
import os 
import datetime
from datetime import datetime
from tqdm import tqdm_notebook as tqdm
from multiprocessing import Pool, TimeoutError, Lock, Value, current_process, cpu_count
from functools import partial

import data_preprocess_function as func
```

```python
Stock = pd.read_csv("D:\\Strategic_Trading\\PriceForecast\\data\\2330\\min_OHLCV.csv", converters = {"ts": str})
Stock["ts"] = pd.to_datetime(Stock["ts"])
Stock
```

```python
df_list = [group[1] for group in Stock.groupby(Stock["ts"].dt.date)]
OHLCV_list = []

if __name__ == '__main__':
    with Pool(processes=12) as pool:
        for i, x in enumerate(tqdm(pool.imap_unordered(func.DailyOHLCV_2330, df_list), total=len(df_list)), 1):
                OHLCV_list.append(x)
```

```python
StockDaily = pd.DataFrame(np.stack([l for l in OHLCV_list]), columns = ['ts', 'open', 'high', 'low', 'close', 'vol', 'vwap'])
StockDaily = StockDaily.sort_values(by='ts').reset_index(drop=True)
StockDaily
```

```python
Index = pd.read_csv('D:\Strategic_Trading\PriceForecast\data\index\min_agg_all.csv', converters = {"ts": str})
Index['ts'] = pd.to_datetime(Index['ts'])
Index
```

```python
df_list = [group[1] for group in Index.groupby(Index["ts"].dt.date)]
OHLCV_list = []

if __name__ == '__main__':
    with Pool(processes=12) as pool:
        for i, x in enumerate(tqdm(pool.imap_unordered(func.DailyOHLCV_Index, df_list), total=len(df_list)), 1):
                OHLCV_list.append(x)
```

```python
IndexDaily = pd.DataFrame(np.stack([l for l in OHLCV_list]), columns = ['ts', 'open_index', 'high_index', 'low_index', 'close_index', 'vol_index'])
IndexDaily = IndexDaily.sort_values(by='ts').reset_index(drop=True)
IndexDaily
```

```python
Daily = pd.merge(StockDaily, IndexDaily, on='ts', how='left')
Daily
```

```python
Daily.to_csv("D:\\Strategic_Trading\\PriceForecast\\data\\2330\\Daily.csv", index=False)
```
