```python
def IntraMonth(row, month_dict):
    y, m, date = row["ts"].year, row["ts"].month, row["ts"].date()
    d = month_dict[y][m]
    total = len(d)
    day = month_dict[y][m][date]
    time = day/(total - 1)
    
    return time
```

```python
import time
tStart = time.time()
data["Intramonth"] = data.apply(IntraMonth, month_dict=month_dict, axis=1)
tEnd = time.time()
print("It cost %f sec" % (tEnd - tStart))
```

```python
def WeekTime(row):
    year = row["ts"].year
    month = row["ts"].month
    date = row["ts"].date()
    week = 0
    while date.month == month:
        week += 1
        date -= timedelta(days=7)
        
    total = len(calendar.monthcalendar(year,month))
    week_time = (week - 1)/(total - 1)

    return week_time
```

```python
import time
tStart = time.time()
data["week"] = data.apply(WeekTime, axis=1)
tEnd = time.time()
print("It cost %f sec" % (tEnd - tStart))
```

```python
def MonthPeriod(row):
    num = row["Intramonth"]
    if num < 0.2:
        return 1
    elif num < 0.4:
        return 2
    elif num < 0.6:
        return 3
    elif num < 0.8:
        return 4
    else:
        return 5
```

```python
data["Period"] = data.apply(MonthPeriod, axis=1)
```

```python
def Cumulative(data):
    data["cumvol"] = data["vol"].cumsum()
    
    return data
```

```python
data = pd.concat(map(Cumulative, [group[1] for group in data.groupby(data["ts"].dt.date)]))
data
```

```python
data["PreCumVol"], data["PreOpen"], data["PreClose"], data["PreHigh"], data["PreLow"] = data["cumvol"].shift(266), data["open"].shift(266), data["close"].shift(266), data["high"].shift(266), data["low"].shift(266)
```

```python
def DailyOHLCV(df):
    
    df['DailyOpen'] = df.iloc[0]['open']
    df['DailyHigh'] = df['high'].max()
    df['DailyLow'] = df['low'].min()
    df['DailyClose'] = df.iloc[-1]['close']
    df['DailyVol'] = df['vol'].sum()
    
    return df
```

```python
data = pd.concat(map(DailyOHLCV, [group[1] for group in data.groupby(data["ts"].dt.date)]))
data
```
